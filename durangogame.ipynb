```python
class Character():
    everyone=[]
    board=4
    
    def __init__(self,name,position):
        self.name=name #name, it could be a player, monster, eggs or door
        self.pos=position # list of two val - [col and row]
                            
    def __del__(self):
        self.pos.clear()
        
    def is_cellbusy(self,step): #check if a cell is free or it is already taken by someone..
                                # in everyone list wich keeps positions of all characters
        if step in self.everyone:
            return True
        else: return False
        
    def monstermove(self):    #random move of the monster
        postemp=[]
        flag=True
        while flag:
            i,j=[random.randint(0,4) for i in range(2)]
            postemp=[i,j]
            if not self.is_cellbusy(postemp): flag=False
            
        self.everyone.remove(self.pos)  #remove old position from all coordinates point list
        self.everyone.append(postemp)  #adding new one
        self.pos=postemp
        return True
    
        
```

```python
import random
from IPython.display import clear_output

def draw_cell(plr,monst,door):
    
    grid=5
    cell=[]
    for i in range(grid):
        row=[]
        for j in range(grid):
            if plr.is_cellbusy([i,j]): 
                if [i,j]==monst.pos: row.append("[M]")
                elif [i,j]==plr.pos: row.append("[P]")
                elif [i,j]==door.pos: row.append("[D]")
                else: row.append("[E]")
            else: row.append("[_]")
        cell.append(row)
        
    for i in range(grid):
        for j in range(grid):
            print(f" {cell[i][j]} ",end='')
        print()

def check_thestep(pos,ch):
    board=4
    
    if ch=='a':
        if (pos[1]-1)<0: return None
        else: pos[1]-=1
    
    if ch=='d':
        if (pos[1]+1)>board: return None
        else: pos[1]+=1
            
    if ch=='w':
        if (pos[0]-1)<0: return None
        else: pos[0]-=1
    
    if ch=='s':
        if (pos[0]+1)>board: return None
        else: pos[0]+=1
            
    return True
    
def open_thedoor(door):
    flag=True
    while flag:
        i=random.randint(0,4)
        board=[0,4]
        if i==0 or i==4: j=random.randint(0,4)
        else: 
            ltemp=random.sample(board,1)
            j=ltemp[0]
        position=[i,j]
        if not door.is_cellbusy(position): flag=False
    
    door.pos=position
    door.everyone.append(position)
    
    return door

def game():
    
    grid=5
    quant_ofeggs=3
#Creating the characters:
    
#the player:
    i,j=[random.randint(0,4) for i in range(2)]
    start_pos=[i,j]
    plr=Character("player",start_pos)
    plr.everyone.append(start_pos)
    
#the monster:
    flag=True
    while flag:
        i,j=[random.randint(0,4) for i in range(2)]
        start_pos=[i,j]
        if not plr.is_cellbusy(start_pos): flag=False
    
    monst=Character("monster",start_pos)    
    monst.everyone.append(start_pos)
#door
    door=Character("door",[])    

#the eggs:
    eggs=[]
    for i in range(quant_ofeggs):
        flag=True
        while flag:
            y,j=[random.randint(0,4) for ii in range(2)]
            start_pos=[y,j]
            if not plr.is_cellbusy(start_pos): flag=False
        eggs.append(Character("egg",start_pos))    
        eggs[i].everyone.append(start_pos)
    

#Start the Game:
    s_enter="Your move... \n('w'-Top 'a'-Left 's'-Down 'd'-Right 'q'-Quit)  "
    ch=''
    allowed_ch=['q','w','a','s','d'] # - list of only acsepted keywords the player can enter
 
    while ch.lower()!='q':
        clear_output()
        draw_cell(plr,monst,door)
        
        ch=input(s_enter)
        
        if ch not in allowed_ch: continue
        
        templist=plr.pos
        if check_thestep(plr.pos,ch):  
            plr.everyone.remove(templist)  #remove old position from all positions list
            plr.everyone.append(plr.pos)   # adding new one
        else: continue
            
        if plr.pos==monst.pos: #the move of player is got to the cell of monster - game over
            #game_over(0)#loose
            print(f"GAME OVER! YOU GOT EATEN BY MONSTER =( ")
            return None
        
        if plr.pos==door.pos: #the move of player is got to door position - Player Win
            print(f"CONGRATS!!! YOU WIN!!!\n")
            return True
        
        for i in range(quant_ofeggs):
            if plr.pos==eggs[i].pos: #the move of player is got to the cell on one of the egg 
                quant_ofeggs-=1
                plr.everyone.remove(plr.pos)
                eggs.pop(i)
                
                clear_output()
                draw_cell(plr,monst,door)
                print(f"You find an egg!... press any key")
                input()
                if quant_ofeggs==0: 
                    #movetonextstep(1)#open the door
                    door=open_thedoor(door)
                    print(f"You collect all the eggs, find the door now .. press any key...")
                    input()
                break
        
        monst.monstermove()
            
```

```python
game()
```
